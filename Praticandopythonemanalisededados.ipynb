```python
import pandas as pd

import numpy as np

import plotly.express as px

import seaborn as sns

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeRegressor
```

```python
data = pd.read_csv("demanda.csv")
```

```python
data = data.dropna()
```

```python
figura = px.scatter(data, x="Unidades vendidas", y="Preço Total",

                 size="Unidades vendidas")

figura.show()
```

```python
print(data.corr())
```

```python
x = data[["Preço Total", "Preço base"]]

y = data["Unidades vendidas"]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y,

                                                test_size=0.2,

                                                random_state=42)

from sklearn.tree import DecisionTreeRegressor

modelo = DecisionTreeRegressor()

modelo.fit(xtrain, ytrain)
```

```python
valores = np.array([[133.00, 140.00]])  #“Preço Total” e “Preço base”

modelo.predict(valores)
```

entregamos valores de preço total e preço base hipotéticos dentro de uma array com nome lógico “valores” para prever quantos produtos serão vendidos usando o comando
